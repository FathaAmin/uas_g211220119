```python
#mengambil dataset
dataset = pd.read_csv('curah_hujan.csv')

#Memebuat Multi bar chart
bulan = dataset['bulan'].values
x = np.arange(len(bulan))
w = 0.3
plt.figure(figsize=(15,5))
plt.bar(x-w, dataset['1981-1990'].values, width=w, label='1981 - 1990', color='b')
plt.bar(x, dataset['1991-2000'].values, width=w, label='1991 - 2000', color='r')
plt.bar(x+w, dataset['2001-2010'].values, width=w, label='2001 - 2010', color='g')

#Mendesain Bar
plt.xticks(x, bulan)
plt.ylim([0,700])
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), fancybox=True, ncol=5)
plt.grid(axis='y')
plt.show()
```

```python
#mengambil dataset
data_suhu = pd.read_csv('suhu_permukaan.csv')

#Memebuat Multi bar chart
waktu = data_suhu['waktu'].values
x = np.arange(len(waktu))
plt.figure(figsize=(10,5))
plt.plot(data_suhu['asimilasi'].values, label='Asimilasi', marker='o', color='b')
plt.plot(data_suhu['observasi'].values, label='Observasi', marker='o', color='y')
plt.plot(data_suhu['non_asimilasi'].values, label='Non_Asimilasi', marker='o', color='r')

#Mendesain Bar
plt.xticks(x, waktu, rotation=50)
plt.ylim([20,27])
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan')
plt.legend(loc='right', bbox_to_anchor=(1, 0.1), fancybox=True, ncol=5)
plt.grid(axis='y')
plt.show()
```
